```python
import os
import json
import yaml
import pickle
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


# ============================================================
# Paths
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System\megambmarketname_n.xls"

OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System"

os.makedirs(OUTPUT_DIR, exist_ok=True)


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_excel(DATASET_PATH)

df.columns = df.columns.str.strip()

required_columns = ["District Name", "Block Name", "Market Name"]

for col in required_columns:
    if col not in df.columns:
        raise ValueError(f"Missing required column: {col}")

df = df.dropna()
df = df.drop_duplicates()

for col in required_columns:
    df[col] = df[col].astype(str).str.strip()


# ============================================================
# Feature Engineering
# ============================================================

df["district_market_count"] = df.groupby("District Name")["Market Name"].transform("count")
df["block_market_count"] = df.groupby("Block Name")["Market Name"].transform("count")

df["market_name_length"] = df["Market Name"].apply(len)
df["block_name_length"] = df["Block Name"].apply(len)
df["district_name_length"] = df["District Name"].apply(len)

df["route_density_score"] = (
    df["district_market_count"] * 0.5 +
    df["block_market_count"] * 0.4 +
    df["market_name_length"] * 0.1
)

df["trade_route_category"] = pd.qcut(
    df["route_density_score"],
    q=3,
    labels=[
        "Low Route Priority",
        "Medium Route Priority",
        "High Route Priority"
    ],
    duplicates="drop"
)


# ============================================================
# Encoding
# ============================================================

label_encoders = {}

for col in required_columns:
    le = LabelEncoder()
    df[col + "_encoded"] = le.fit_transform(df[col])
    label_encoders[col] = le

target_encoder = LabelEncoder()
df["target"] = target_encoder.fit_transform(df["trade_route_category"])


# ============================================================
# Features and Target
# ============================================================

feature_cols = [
    "District Name_encoded",
    "Block Name_encoded",
    "Market Name_encoded",
    "district_market_count",
    "block_market_count",
    "market_name_length",
    "block_name_length",
    "district_name_length",
    "route_density_score"
]

X = df[feature_cols]
y = df["target"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# ============================================================
# Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)


# ============================================================
# ML Models
# ============================================================

models = {
    "Random Forest": RandomForestClassifier(
        n_estimators=200,
        max_depth=8,
        random_state=42
    ),
    "Decision Tree": DecisionTreeClassifier(
        random_state=42,
        max_depth=6
    ),
    "Gradient Boosting": GradientBoostingClassifier(
        random_state=42
    )
}

results = []
trained_models = {}

for model_name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    acc = accuracy_score(y_test, pred)

    results.append({
        "Model": model_name,
        "Accuracy": acc
    })

    trained_models[model_name] = model


# ============================================================
# Deep Learning H5 Model
# ============================================================

num_classes = len(np.unique(y))

y_train_cat = to_categorical(y_train, num_classes)
y_test_cat = to_categorical(y_test, num_classes)

h5_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

h5_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = h5_model.fit(
    X_train,
    y_train_cat,
    validation_data=(X_test, y_test_cat),
    epochs=50,
    batch_size=8,
    verbose=1
)

h5_loss, h5_accuracy = h5_model.evaluate(X_test, y_test_cat, verbose=0)

results.append({
    "Model": "Deep Learning H5",
    "Accuracy": h5_accuracy
})


# ============================================================
# Result CSV
# ============================================================

result_df = pd.DataFrame(results)

result_csv_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_result.csv"
)

result_df.to_csv(result_csv_path, index=False)

print("\nResult CSV:")
print(result_df)


# ============================================================
# Best Model
# ============================================================

best_model_name = result_df.sort_values(
    by="Accuracy",
    ascending=False
).iloc[0]["Model"]

if best_model_name == "Deep Learning H5":
    best_model = None
    final_pred_encoded = np.argmax(h5_model.predict(X_scaled), axis=1)
else:
    best_model = trained_models[best_model_name]
    final_pred_encoded = best_model.predict(X_scaled)

df["predicted_route_category"] = target_encoder.inverse_transform(
    final_pred_encoded
)


# ============================================================
# Prediction CSV
# ============================================================

prediction_csv_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_prediction.csv"
)

df.to_csv(prediction_csv_path, index=False)

print("\nPrediction CSV Saved:")
print(prediction_csv_path)


# ============================================================
# Graph 1: Accuracy Graph
# ============================================================

accuracy_graph_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_accuracy_graph.png"
)

plt.figure(figsize=(8, 5))
plt.bar(result_df["Model"], result_df["Accuracy"])
plt.title("Model Accuracy Graph")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.xticks(rotation=25)
plt.tight_layout()
plt.savefig(accuracy_graph_path)
plt.show()


# ============================================================
# Graph 2: Comparison Graph
# ============================================================

comparison_graph_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_comparison_graph.png"
)

plt.figure(figsize=(8, 5))
plt.plot(
    result_df["Model"],
    result_df["Accuracy"],
    marker="o"
)
plt.title("Model Comparison Graph")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.xticks(rotation=25)
plt.grid(True)
plt.tight_layout()
plt.savefig(comparison_graph_path)
plt.show()


# ============================================================
# Graph 3: Heatmap
# ============================================================

heatmap_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_heatmap.png"
)

corr = df[feature_cols + ["target"]].corr()

plt.figure(figsize=(10, 7))
plt.imshow(corr, aspect="auto")
plt.colorbar()
plt.xticks(
    range(len(corr.columns)),
    corr.columns,
    rotation=90
)
plt.yticks(
    range(len(corr.columns)),
    corr.columns
)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_path)
plt.show()


# ============================================================
# Graph 4: Result Graph
# ============================================================

result_graph_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_result_graph.png"
)

plt.figure(figsize=(8, 5))
plt.barh(result_df["Model"], result_df["Accuracy"])
plt.title("Trade Route Result Graph")
plt.xlabel("Accuracy")
plt.ylabel("Model")
plt.tight_layout()
plt.savefig(result_graph_path)
plt.show()


# ============================================================
# Graph 5: Prediction Graph
# ============================================================

prediction_graph_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_prediction_graph.png"
)

prediction_counts = df["predicted_route_category"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(prediction_counts.index, prediction_counts.values)
plt.title("Predicted Trade Route Category Distribution")
plt.xlabel("Predicted Category")
plt.ylabel("Count")
plt.xticks(rotation=25)
plt.tight_layout()
plt.savefig(prediction_graph_path)
plt.show()


# ============================================================
# Graph 6: Training Accuracy Graph
# ============================================================

training_accuracy_graph_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_training_accuracy_graph.png"
)

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("H5 Training Accuracy Graph")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(training_accuracy_graph_path)
plt.show()


# ============================================================
# Confusion Matrix Graph
# ============================================================

if best_model_name == "Deep Learning H5":
    test_pred = np.argmax(h5_model.predict(X_test), axis=1)
else:
    test_pred = trained_models[best_model_name].predict(X_test)

cm = confusion_matrix(y_test, test_pred)

confusion_matrix_graph_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_confusion_matrix.png"
)

plt.figure(figsize=(7, 5))
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.colorbar()

classes = target_encoder.classes_

plt.xticks(np.arange(len(classes)), classes, rotation=45)
plt.yticks(np.arange(len(classes)), classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.savefig(confusion_matrix_graph_path)
plt.show()


# ============================================================
# Save PKL Files
# ============================================================

rf_model_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_rf_model.pkl"
)

scaler_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_scaler.pkl"
)

encoder_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_label_encoders.pkl"
)

with open(rf_model_path, "wb") as f:
    pickle.dump(trained_models["Random Forest"], f)

with open(scaler_path, "wb") as f:
    pickle.dump(scaler, f)

with open(encoder_path, "wb") as f:
    pickle.dump(
        {
            "feature_encoders": label_encoders,
            "target_encoder": target_encoder
        },
        f
    )


# ============================================================
# Save H5 Model
# ============================================================

h5_model_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_optimization_model.h5"
)

h5_model.save(h5_model_path)


# ============================================================
# Save YAML Config
# ============================================================

yaml_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_config.yaml"
)

config = {
    "project_name": "Agricultural Trade Route Optimization System",
    "author": "Sagnik Patra",
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "target_column": "trade_route_category",
    "features": feature_cols,
    "best_model": str(best_model_name),
    "files_generated": {
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "h5_model": h5_model_path,
        "rf_model_pkl": rf_model_path,
        "scaler_pkl": scaler_path,
        "encoder_pkl": encoder_path,
        "json_results": "trade_route_results.json",
        "accuracy_graph": accuracy_graph_path,
        "comparison_graph": comparison_graph_path,
        "heatmap": heatmap_path,
        "result_graph": result_graph_path,
        "prediction_graph": prediction_graph_path,
        "training_accuracy_graph": training_accuracy_graph_path,
        "confusion_matrix_graph": confusion_matrix_graph_path
    }
}

with open(yaml_path, "w") as f:
    yaml.dump(config, f, default_flow_style=False)


# ============================================================
# Save JSON Results
# ============================================================

json_path = os.path.join(
    OUTPUT_DIR,
    "trade_route_results.json"
)

json_results = {
    "project_name": "Agricultural Trade Route Optimization System",
    "author": "Sagnik Patra",
    "best_model": str(best_model_name),
    "model_results": result_df.to_dict(orient="records"),
    "classification_report": classification_report(
        y_test,
        test_pred,
        target_names=target_encoder.classes_,
        output_dict=True
    ),
    "confusion_matrix": cm.tolist(),
    "total_records": int(len(df)),
    "total_features": int(len(feature_cols)),
    "target_classes": list(target_encoder.classes_),
    "saved_files": {
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "h5_model": h5_model_path,
        "pkl_model": rf_model_path,
        "pkl_scaler": scaler_path,
        "pkl_encoders": encoder_path,
        "yaml_config": yaml_path,
        "accuracy_graph": accuracy_graph_path,
        "comparison_graph": comparison_graph_path,
        "heatmap": heatmap_path,
        "result_graph": result_graph_path,
        "prediction_graph": prediction_graph_path,
        "training_accuracy_graph": training_accuracy_graph_path,
        "confusion_matrix_graph": confusion_matrix_graph_path
    }
}

with open(json_path, "w") as f:
    json.dump(json_results, f, indent=4)


# ============================================================
# Final Output
# ============================================================

print("\nAll Files Saved Successfully")
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("H5 Model:", h5_model_path)
print("PKL Model:", rf_model_path)
print("Scaler:", scaler_path)
print("Encoders:", encoder_path)
print("YAML:", yaml_path)
print("JSON:", json_path)

print("\nGraphs Saved:")
print("Accuracy Graph:", accuracy_graph_path)
print("Comparison Graph:", comparison_graph_path)
print("Heatmap:", heatmap_path)
print("Result Graph:", result_graph_path)
print("Prediction Graph:", prediction_graph_path)
print("Training Accuracy Graph:", training_accuracy_graph_path)
print("Confusion Matrix Graph:", confusion_matrix_graph_path)

print("\nBest Model:", best_model_name)
```
