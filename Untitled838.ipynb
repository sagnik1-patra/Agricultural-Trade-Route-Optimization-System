```python
import os
import json
import yaml
import pickle
import warnings
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


# ============================================================
# Paths
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System\megambmarketname_n.xls"

OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System"

os.makedirs(OUTPUT_DIR, exist_ok=True)


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_excel(DATASET_PATH)

print("Dataset Loaded Successfully")
print(df.head())
print(df.info())


# ============================================================
# Clean Columns
# ============================================================

df.columns = df.columns.str.strip()

required_columns = ["District Name", "Block Name", "Market Name"]

for col in required_columns:
    if col not in df.columns:
        raise ValueError(f"Missing required column: {col}")

df = df.dropna()
df = df.drop_duplicates()

df["District Name"] = df["District Name"].astype(str).str.strip()
df["Block Name"] = df["Block Name"].astype(str).str.strip()
df["Market Name"] = df["Market Name"].astype(str).str.strip()


# ============================================================
# Feature Engineering
# ============================================================

district_market_count = df.groupby("District Name")["Market Name"].transform("count")
block_market_count = df.groupby("Block Name")["Market Name"].transform("count")

df["district_market_count"] = district_market_count
df["block_market_count"] = block_market_count

df["market_name_length"] = df["Market Name"].apply(len)
df["block_name_length"] = df["Block Name"].apply(len)
df["district_name_length"] = df["District Name"].apply(len)

df["route_density_score"] = (
    df["district_market_count"] * 0.5 +
    df["block_market_count"] * 0.4 +
    df["market_name_length"] * 0.1
)

# Synthetic target for route optimization category
df["trade_route_category"] = pd.qcut(
    df["route_density_score"],
    q=3,
    labels=["Low Route Priority", "Medium Route Priority", "High Route Priority"]
)

print("\nFeature Engineered Data:")
print(df.head())


# ============================================================
# Encoding
# ============================================================

label_encoders = {}

categorical_cols = ["District Name", "Block Name", "Market Name"]

for col in categorical_cols:
    le = LabelEncoder()
    df[col + "_encoded"] = le.fit_transform(df[col])
    label_encoders[col] = le

target_encoder = LabelEncoder()
df["target"] = target_encoder.fit_transform(df["trade_route_category"])


# ============================================================
# Features and Target
# ============================================================

feature_cols = [
    "District Name_encoded",
    "Block Name_encoded",
    "Market Name_encoded",
    "district_market_count",
    "block_market_count",
    "market_name_length",
    "block_name_length",
    "district_name_length",
    "route_density_score"
]

X = df[feature_cols]
y = df["target"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# ============================================================
# Train Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)


# ============================================================
# Random Forest Model
# ============================================================

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=8
)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)

rf_accuracy = accuracy_score(y_test, rf_pred)

print("\nRandom Forest Accuracy:", rf_accuracy)


# ============================================================
# Deep Learning H5 Model
# ============================================================

num_classes = len(np.unique(y))

y_train_cat = to_categorical(y_train, num_classes)
y_test_cat = to_categorical(y_test, num_classes)

h5_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

h5_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = h5_model.fit(
    X_train,
    y_train_cat,
    validation_data=(X_test, y_test_cat),
    epochs=50,
    batch_size=8,
    verbose=1
)

h5_loss, h5_accuracy = h5_model.evaluate(X_test, y_test_cat, verbose=0)

print("\nH5 Model Accuracy:", h5_accuracy)


# ============================================================
# Predictions
# ============================================================

df["predicted_route_category"] = target_encoder.inverse_transform(
    rf_model.predict(X_scaled)
)

prediction_path = os.path.join(OUTPUT_DIR, "trade_route_predictions.csv")
df.to_csv(prediction_path, index=False)


# ============================================================
# Save PKL Files
# ============================================================

pkl_model_path = os.path.join(OUTPUT_DIR, "trade_route_rf_model.pkl")
pkl_scaler_path = os.path.join(OUTPUT_DIR, "trade_route_scaler.pkl")
pkl_encoder_path = os.path.join(OUTPUT_DIR, "trade_route_label_encoders.pkl")

with open(pkl_model_path, "wb") as f:
    pickle.dump(rf_model, f)

with open(pkl_scaler_path, "wb") as f:
    pickle.dump(scaler, f)

with open(pkl_encoder_path, "wb") as f:
    pickle.dump(
        {
            "feature_encoders": label_encoders,
            "target_encoder": target_encoder
        },
        f
    )


# ============================================================
# Save H5 Model
# ============================================================

h5_model_path = os.path.join(OUTPUT_DIR, "trade_route_optimization_model.h5")
h5_model.save(h5_model_path)


# ============================================================
# Save YAML Config
# ============================================================

config = {
    "project_name": "Agricultural Trade Route Optimization System",
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "features": feature_cols,
    "target_column": "trade_route_category",
    "models": {
        "machine_learning_model": "RandomForestClassifier",
        "deep_learning_model": "Keras Sequential Neural Network"
    },
    "outputs": {
        "h5_model": h5_model_path,
        "pkl_model": pkl_model_path,
        "pkl_scaler": pkl_scaler_path,
        "pkl_encoders": pkl_encoder_path,
        "prediction_csv": prediction_path
    }
}

yaml_path = os.path.join(OUTPUT_DIR, "trade_route_config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(config, f, default_flow_style=False)


# ============================================================
# Save JSON Results
# ============================================================

results = {
    "random_forest_accuracy": float(rf_accuracy),
    "h5_model_accuracy": float(h5_accuracy),
    "classification_report": classification_report(
        y_test,
        rf_pred,
        target_names=target_encoder.classes_,
        output_dict=True
    ),
    "confusion_matrix": confusion_matrix(y_test, rf_pred).tolist(),
    "total_records": int(len(df)),
    "total_features": int(len(feature_cols)),
    "target_classes": list(target_encoder.classes_)
}

json_path = os.path.join(OUTPUT_DIR, "trade_route_results.json")

with open(json_path, "w") as f:
    json.dump(results, f, indent=4)


# ============================================================
# Final Output
# ============================================================

print("\nAll Files Saved Successfully:")
print("H5 Model:", h5_model_path)
print("PKL Model:", pkl_model_path)
print("PKL Scaler:", pkl_scaler_path)
print("PKL Encoders:", pkl_encoder_path)
print("YAML Config:", yaml_path)
print("JSON Results:", json_path)
print("Prediction CSV:", prediction_path)
```
