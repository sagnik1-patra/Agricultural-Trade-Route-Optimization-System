```python
import os
import json
import yaml
import pickle
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


# ============================================================
# Paths
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System\megambmarketname_n.xls"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "aloa_"


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_excel(DATASET_PATH)
df.columns = df.columns.str.strip()

required_columns = ["District Name", "Block Name", "Market Name"]

for col in required_columns:
    if col not in df.columns:
        raise ValueError(f"Missing required column: {col}")

df = df.dropna()
df = df.drop_duplicates()

for col in required_columns:
    df[col] = df[col].astype(str).str.strip()


# ============================================================
# Feature Engineering
# ============================================================

df["district_market_count"] = df.groupby("District Name")["Market Name"].transform("count")
df["block_market_count"] = df.groupby("Block Name")["Market Name"].transform("count")

df["market_name_length"] = df["Market Name"].apply(len)
df["block_name_length"] = df["Block Name"].apply(len)
df["district_name_length"] = df["District Name"].apply(len)

df["route_density_score"] = (
    df["district_market_count"] * 0.5 +
    df["block_market_count"] * 0.4 +
    df["market_name_length"] * 0.1
)

df["trade_route_category"] = pd.qcut(
    df["route_density_score"],
    q=3,
    labels=[
        "Low Route Priority",
        "Medium Route Priority",
        "High Route Priority"
    ],
    duplicates="drop"
)


# ============================================================
# Encoding
# ============================================================

label_encoders = {}

for col in required_columns:
    le = LabelEncoder()
    df[col + "_encoded"] = le.fit_transform(df[col])
    label_encoders[col] = le

target_encoder = LabelEncoder()
df["target"] = target_encoder.fit_transform(df["trade_route_category"])


# ============================================================
# Features and Target
# ============================================================

feature_cols = [
    "District Name_encoded",
    "Block Name_encoded",
    "Market Name_encoded",
    "district_market_count",
    "block_market_count",
    "market_name_length",
    "block_name_length",
    "district_name_length",
    "route_density_score"
]

X = df[feature_cols].values
y = df["target"].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# ============================================================
# ALOA Feature Selection
# Ant Lion Optimization Algorithm
# ============================================================

def fitness_function(mask, X_data, y_data):
    selected_features = np.where(mask == 1)[0]

    if len(selected_features) == 0:
        return 0

    X_selected = X_data[:, selected_features]

    X_train_fs, X_test_fs, y_train_fs, y_test_fs = train_test_split(
        X_selected,
        y_data,
        test_size=0.2,
        random_state=42,
        stratify=y_data
    )

    model = RandomForestClassifier(
        n_estimators=100,
        max_depth=8,
        random_state=42
    )

    model.fit(X_train_fs, y_train_fs)
    pred = model.predict(X_test_fs)

    accuracy = accuracy_score(y_test_fs, pred)
    feature_penalty = len(selected_features) / X_data.shape[1]

    return accuracy - (0.01 * feature_penalty)


def roulette_wheel_selection(fitness_values):
    fitness_values = np.array(fitness_values)
    min_fit = np.min(fitness_values)

    if min_fit < 0:
        fitness_values = fitness_values - min_fit + 1e-10

    total_fitness = np.sum(fitness_values)

    if total_fitness == 0:
        return np.random.randint(0, len(fitness_values))

    probabilities = fitness_values / total_fitness
    return np.random.choice(len(fitness_values), p=probabilities)


def binary_random_walk(num_features, generation, max_generations):
    random_steps = np.where(np.random.rand(num_features) > 0.5, 1, -1)
    walk = np.cumsum(random_steps)

    min_walk = np.min(walk)
    max_walk = np.max(walk)

    if max_walk - min_walk == 0:
        normalized_walk = np.random.rand(num_features)
    else:
        normalized_walk = (walk - min_walk) / (max_walk - min_walk)

    threshold = 0.5 + 0.3 * (generation / max_generations)
    return (normalized_walk > threshold).astype(int)


def aloa_feature_selection(
    X_data,
    y_data,
    population_size=20,
    generations=25
):
    num_features = X_data.shape[1]

    ants = np.random.randint(0, 2, size=(population_size, num_features))
    antlions = np.random.randint(0, 2, size=(population_size, num_features))

    for i in range(population_size):
        if np.sum(ants[i]) == 0:
            ants[i, np.random.randint(0, num_features)] = 1
        if np.sum(antlions[i]) == 0:
            antlions[i, np.random.randint(0, num_features)] = 1

    antlion_fitness = np.array([
        fitness_function(antlion, X_data, y_data)
        for antlion in antlions
    ])

    elite_index = np.argmax(antlion_fitness)
    elite_antlion = antlions[elite_index].copy()
    elite_fitness = antlion_fitness[elite_index]

    fitness_history = []

    for generation in range(generations):
        new_ants = []

        for i in range(population_size):
            selected_index = roulette_wheel_selection(antlion_fitness)
            selected_antlion = antlions[selected_index]

            walk_selected = binary_random_walk(
                num_features,
                generation,
                generations
            )

            walk_elite = binary_random_walk(
                num_features,
                generation,
                generations
            )

            candidate = np.zeros(num_features, dtype=int)

            for j in range(num_features):
                if np.random.rand() < 0.5:
                    candidate[j] = selected_antlion[j]
                else:
                    candidate[j] = elite_antlion[j]

                if np.random.rand() < 0.3:
                    candidate[j] = walk_selected[j]

                if np.random.rand() < 0.3:
                    candidate[j] = walk_elite[j]

            mutation_rate = 0.2 * (1 - generation / generations)
            mutation_mask = np.random.rand(num_features) < mutation_rate
            candidate[mutation_mask] = 1 - candidate[mutation_mask]

            if np.sum(candidate) == 0:
                candidate[np.random.randint(0, num_features)] = 1

            new_ants.append(candidate)

        ants = np.array(new_ants)

        ant_fitness = np.array([
            fitness_function(ant, X_data, y_data)
            for ant in ants
        ])

        combined_population = np.vstack((antlions, ants))
        combined_fitness = np.concatenate((antlion_fitness, ant_fitness))

        sorted_indices = np.argsort(combined_fitness)[::-1]

        antlions = combined_population[sorted_indices[:population_size]]
        antlion_fitness = combined_fitness[sorted_indices[:population_size]]

        if antlion_fitness[0] > elite_fitness:
            elite_fitness = antlion_fitness[0]
            elite_antlion = antlions[0].copy()

        fitness_history.append(elite_fitness)

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best ALOA Fitness: {elite_fitness:.4f}"
        )

    return elite_antlion, elite_fitness, fitness_history


selected_mask, best_aloa_fitness, aloa_history = aloa_feature_selection(
    X_scaled,
    y
)

selected_feature_indices = np.where(selected_mask == 1)[0]
selected_features = [feature_cols[i] for i in selected_feature_indices]

print("\nSelected Features by ALOA:")
print(selected_features)

X_selected = X_scaled[:, selected_feature_indices]


# ============================================================
# Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_selected,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)


# ============================================================
# ALOA Optimized Random Forest Model
# ============================================================

aloa_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    random_state=42
)

aloa_model.fit(X_train, y_train)

y_pred = aloa_model.predict(X_test)
aloa_accuracy = accuracy_score(y_test, y_pred)

print("\nALOA Optimized Random Forest Accuracy:", aloa_accuracy)


# ============================================================
# H5 Deep Learning Model
# ============================================================

num_classes = len(np.unique(y))

y_train_cat = to_categorical(y_train, num_classes)
y_test_cat = to_categorical(y_test, num_classes)

h5_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

h5_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = h5_model.fit(
    X_train,
    y_train_cat,
    validation_data=(X_test, y_test_cat),
    epochs=50,
    batch_size=8,
    verbose=1
)

h5_loss, h5_accuracy = h5_model.evaluate(X_test, y_test_cat, verbose=0)


# ============================================================
# Result CSV
# ============================================================

result_df = pd.DataFrame([
    {
        "Model": "ALOA Optimized Random Forest",
        "Accuracy": aloa_accuracy,
        "Fitness Score": best_aloa_fitness,
        "Selected Features": ", ".join(selected_features)
    },
    {
        "Model": "ALOA Optimized H5 Neural Network",
        "Accuracy": h5_accuracy,
        "Fitness Score": best_aloa_fitness,
        "Selected Features": ", ".join(selected_features)
    }
])

result_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "result.csv")
result_df.to_csv(result_csv_path, index=False)

print("\nResult CSV:")
print(result_df)


# ============================================================
# Prediction CSV
# ============================================================

final_pred_encoded = aloa_model.predict(X_selected)

df["predicted_route_category"] = target_encoder.inverse_transform(
    final_pred_encoded
)

prediction_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv")
df.to_csv(prediction_csv_path, index=False)

print("\nPrediction CSV Saved:", prediction_csv_path)


# ============================================================
# Graphs
# ============================================================

accuracy_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")
plt.figure(figsize=(8, 5))
plt.bar(result_df["Model"], result_df["Accuracy"])
plt.title("ALOA Model Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(accuracy_graph_path)
plt.show()

comparison_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")
plt.figure(figsize=(8, 5))
plt.plot(result_df["Model"], result_df["Accuracy"], marker="o")
plt.title("ALOA Model Comparison Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.grid(True)
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(comparison_graph_path)
plt.show()

heatmap_path = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")
corr_df = pd.DataFrame(X_selected, columns=selected_features)
corr_df["target"] = y
corr = corr_df.corr()

plt.figure(figsize=(10, 7))
plt.imshow(corr, aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("ALOA Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_path)
plt.show()

result_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")
plt.figure(figsize=(8, 5))
plt.barh(result_df["Model"], result_df["Accuracy"])
plt.title("ALOA Result Graph")
plt.xlabel("Accuracy")
plt.ylabel("Model")
plt.tight_layout()
plt.savefig(result_graph_path)
plt.show()

prediction_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")
prediction_counts = df["predicted_route_category"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(prediction_counts.index, prediction_counts.values)
plt.title("ALOA Prediction Category Distribution")
plt.xlabel("Predicted Route Category")
plt.ylabel("Count")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(prediction_graph_path)
plt.show()

optimization_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "optimization_graph.png")
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(aloa_history) + 1), aloa_history, marker="o")
plt.title("ALOA Optimization Progress Graph")
plt.xlabel("Generation")
plt.ylabel("Best Fitness Score")
plt.grid(True)
plt.tight_layout()
plt.savefig(optimization_graph_path)
plt.show()

training_accuracy_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "training_accuracy_graph.png"
)

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("ALOA H5 Training Accuracy Graph")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(training_accuracy_graph_path)
plt.show()

cm = confusion_matrix(y_test, y_pred)

confusion_matrix_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "confusion_matrix_graph.png"
)

plt.figure(figsize=(7, 5))
plt.imshow(cm, interpolation="nearest")
plt.title("ALOA Confusion Matrix")
plt.colorbar()

classes = target_encoder.classes_

plt.xticks(np.arange(len(classes)), classes, rotation=45)
plt.yticks(np.arange(len(classes)), classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.savefig(confusion_matrix_graph_path)
plt.show()


# ============================================================
# Save PKL Files
# ============================================================

model_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "model.pkl")
scaler_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "encoders.pkl")
feature_selection_pkl_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "selected_features.pkl"
)

with open(model_pkl_path, "wb") as f:
    pickle.dump(aloa_model, f)

with open(scaler_pkl_path, "wb") as f:
    pickle.dump(scaler, f)

with open(encoder_pkl_path, "wb") as f:
    pickle.dump(
        {
            "feature_encoders": label_encoders,
            "target_encoder": target_encoder
        },
        f
    )

with open(feature_selection_pkl_path, "wb") as f:
    pickle.dump(
        {
            "selected_mask": selected_mask,
            "selected_features": selected_features,
            "best_fitness": best_aloa_fitness
        },
        f
    )


# ============================================================
# Save H5 Model
# ============================================================

h5_model_path = os.path.join(OUTPUT_DIR, PREFIX + "model.h5")
h5_model.save(h5_model_path)


# ============================================================
# Save YAML Config
# ============================================================

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

config = {
    "project_name": "Agricultural Trade Route Optimization System",
    "author": "Sagnik Patra",
    "algorithm": "Ant Lion Optimization Algorithm",
    "prefix": PREFIX,
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "target_column": "trade_route_category",
    "all_features": feature_cols,
    "selected_features": selected_features,
    "aloa_parameters": {
        "population_size": 20,
        "generations": 25,
        "best_fitness": float(best_aloa_fitness)
    },
    "files_generated": {
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "h5_model": h5_model_path,
        "pkl_model": model_pkl_path,
        "pkl_scaler": scaler_pkl_path,
        "pkl_encoders": encoder_pkl_path,
        "pkl_selected_features": feature_selection_pkl_path,
        "accuracy_graph": accuracy_graph_path,
        "comparison_graph": comparison_graph_path,
        "heatmap": heatmap_path,
        "result_graph": result_graph_path,
        "prediction_graph": prediction_graph_path,
        "optimization_graph": optimization_graph_path,
        "training_accuracy_graph": training_accuracy_graph_path,
        "confusion_matrix_graph": confusion_matrix_graph_path
    }
}

with open(yaml_path, "w") as f:
    yaml.dump(config, f, default_flow_style=False)


# ============================================================
# Save JSON Results
# ============================================================

json_path = os.path.join(OUTPUT_DIR, PREFIX + "results.json")

json_results = {
    "project_name": "Agricultural Trade Route Optimization System",
    "author": "Sagnik Patra",
    "algorithm": "Ant Lion Optimization Algorithm",
    "prefix": PREFIX,
    "aloa_accuracy": float(aloa_accuracy),
    "h5_accuracy": float(h5_accuracy),
    "best_aloa_fitness": float(best_aloa_fitness),
    "selected_features": selected_features,
    "model_results": result_df.to_dict(orient="records"),
    "classification_report": classification_report(
        y_test,
        y_pred,
        target_names=target_encoder.classes_,
        output_dict=True
    ),
    "confusion_matrix": cm.tolist(),
    "total_records": int(len(df)),
    "total_selected_features": int(len(selected_features)),
    "target_classes": list(target_encoder.classes_),
    "saved_files": {
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "h5_model": h5_model_path,
        "pkl_model": model_pkl_path,
        "pkl_scaler": scaler_pkl_path,
        "pkl_encoders": encoder_pkl_path,
        "pkl_selected_features": feature_selection_pkl_path,
        "yaml_config": yaml_path,
        "accuracy_graph": accuracy_graph_path,
        "comparison_graph": comparison_graph_path,
        "heatmap": heatmap_path,
        "result_graph": result_graph_path,
        "prediction_graph": prediction_graph_path,
        "optimization_graph": optimization_graph_path,
        "training_accuracy_graph": training_accuracy_graph_path,
        "confusion_matrix_graph": confusion_matrix_graph_path
    }
}

with open(json_path, "w") as f:
    json.dump(json_results, f, indent=4)


# ============================================================
# Final Output
# ============================================================

print("\nAll ALOA Files Saved Successfully")

print("\nCSV Files:")
print(result_csv_path)
print(prediction_csv_path)

print("\nModel Files:")
print(h5_model_path)
print(model_pkl_path)
print(scaler_pkl_path)
print(encoder_pkl_path)
print(feature_selection_pkl_path)

print("\nConfig Files:")
print(yaml_path)
print(json_path)

print("\nGraph Files:")
print(accuracy_graph_path)
print(comparison_graph_path)
print(heatmap_path)
print(result_graph_path)
print(prediction_graph_path)
print(optimization_graph_path)
print(training_accuracy_graph_path)
print(confusion_matrix_graph_path)

print("\nSelected Features:")
print(selected_features)

print("\nALOA Accuracy:", aloa_accuracy)
print("H5 Accuracy:", h5_accuracy)
print("Best ALOA Fitness:", best_aloa_fitness)
```
