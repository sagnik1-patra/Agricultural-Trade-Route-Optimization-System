```python
import os
import json
import yaml
import pickle
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System\megambmarketname_n.xls"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "psa_"


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_excel(DATASET_PATH)
df.columns = df.columns.str.strip()

required_columns = ["District Name", "Block Name", "Market Name"]

for col in required_columns:
    if col not in df.columns:
        raise ValueError(f"Missing required column: {col}")

df = df.dropna()
df = df.drop_duplicates()

for col in required_columns:
    df[col] = df[col].astype(str).str.strip()


# ============================================================
# Feature Engineering
# ============================================================

df["district_market_count"] = df.groupby("District Name")["Market Name"].transform("count")
df["block_market_count"] = df.groupby("Block Name")["Market Name"].transform("count")

df["market_name_length"] = df["Market Name"].apply(len)
df["block_name_length"] = df["Block Name"].apply(len)
df["district_name_length"] = df["District Name"].apply(len)

df["route_density_score"] = (
    df["district_market_count"] * 0.5 +
    df["block_market_count"] * 0.4 +
    df["market_name_length"] * 0.1
)

df["trade_route_category"] = pd.qcut(
    df["route_density_score"],
    q=3,
    labels=[
        "Low Route Priority",
        "Medium Route Priority",
        "High Route Priority"
    ],
    duplicates="drop"
)


# ============================================================
# Encoding
# ============================================================

label_encoders = {}

for col in required_columns:
    le = LabelEncoder()
    df[col + "_encoded"] = le.fit_transform(df[col])
    label_encoders[col] = le

target_encoder = LabelEncoder()
df["target"] = target_encoder.fit_transform(df["trade_route_category"])


# ============================================================
# Features and Target
# ============================================================

feature_cols = [
    "District Name_encoded",
    "Block Name_encoded",
    "Market Name_encoded",
    "district_market_count",
    "block_market_count",
    "market_name_length",
    "block_name_length",
    "district_name_length",
    "route_density_score"
]

X = df[feature_cols].values
y = df["target"].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# ============================================================
# Fitness Function
# ============================================================

def fitness_function(mask, X_data, y_data):
    selected_features = np.where(mask == 1)[0]

    if len(selected_features) == 0:
        return 0

    X_selected = X_data[:, selected_features]

    X_train_fs, X_test_fs, y_train_fs, y_test_fs = train_test_split(
        X_selected,
        y_data,
        test_size=0.2,
        random_state=42,
        stratify=y_data
    )

    model = RandomForestClassifier(
        n_estimators=100,
        max_depth=8,
        random_state=42
    )

    model.fit(X_train_fs, y_train_fs)
    pred = model.predict(X_test_fs)

    accuracy = accuracy_score(y_test_fs, pred)
    feature_penalty = len(selected_features) / X_data.shape[1]

    return accuracy - (0.01 * feature_penalty)


# ============================================================
# Hybrid CSA + PSO Feature Selection
# ============================================================

def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def hybrid_csa_pso_feature_selection(
    X_data,
    y_data,
    population_size=20,
    generations=25,
    clone_rate=5,
    mutation_rate=0.15,
    replacement_rate=0.3,
    w=0.7,
    c1=1.5,
    c2=1.5
):
    num_features = X_data.shape[1]

    population = np.random.randint(
        0,
        2,
        size=(population_size, num_features)
    )

    velocity = np.random.uniform(
        -1,
        1,
        size=(population_size, num_features)
    )

    for i in range(population_size):
        if np.sum(population[i]) == 0:
            population[i, np.random.randint(0, num_features)] = 1

    personal_best_position = population.copy()

    personal_best_score = np.array([
        fitness_function(individual, X_data, y_data)
        for individual in population
    ])

    global_best_index = np.argmax(personal_best_score)
    global_best_position = personal_best_position[global_best_index].copy()
    global_best_score = personal_best_score[global_best_index]

    fitness_history = []

    for generation in range(generations):

        # ---------------- CSA: Cloning + Hypermutation ----------------

        fitness_scores = np.array([
            fitness_function(individual, X_data, y_data)
            for individual in population
        ])

        sorted_indices = np.argsort(fitness_scores)[::-1]
        population = population[sorted_indices]
        fitness_scores = fitness_scores[sorted_indices]

        elite_count = population_size // 2
        elites = population[:elite_count]

        clones = []

        for rank, antibody in enumerate(elites):
            num_clones = max(1, clone_rate * (elite_count - rank))

            for _ in range(num_clones):
                clone = antibody.copy()

                adaptive_mutation = mutation_rate * (1 - fitness_scores[rank])

                mutation_mask = np.random.rand(num_features) < adaptive_mutation
                clone[mutation_mask] = 1 - clone[mutation_mask]

                if np.sum(clone) == 0:
                    clone[np.random.randint(0, num_features)] = 1

                clones.append(clone)

        clone_fitness = np.array([
            fitness_function(clone, X_data, y_data)
            for clone in clones
        ])

        best_clone_indices = np.argsort(clone_fitness)[::-1]
        best_clones = np.array(clones)[best_clone_indices[:elite_count]]

        replace_count = int(population_size * replacement_rate)

        random_population = np.random.randint(
            0,
            2,
            size=(replace_count, num_features)
        )

        for i in range(replace_count):
            if np.sum(random_population[i]) == 0:
                random_population[i, np.random.randint(0, num_features)] = 1

        remaining_count = population_size - elite_count - replace_count

        if remaining_count > 0:
            population = np.vstack([
                best_clones,
                population[:remaining_count],
                random_population
            ])
        else:
            population = np.vstack([
                best_clones,
                random_population
            ])[:population_size]

        if population.shape[0] < population_size:
            extra = np.random.randint(
                0,
                2,
                size=(population_size - population.shape[0], num_features)
            )
            population = np.vstack([population, extra])

        # ---------------- PSO Update ----------------

        for i in range(population_size):
            r1 = np.random.rand(num_features)
            r2 = np.random.rand(num_features)

            velocity[i] = (
                w * velocity[i]
                + c1 * r1 * (personal_best_position[i] - population[i])
                + c2 * r2 * (global_best_position - population[i])
            )

            probability = sigmoid(velocity[i])

            pso_candidate = (
                np.random.rand(num_features) < probability
            ).astype(int)

            if np.sum(pso_candidate) == 0:
                pso_candidate[np.random.randint(0, num_features)] = 1

            pso_score = fitness_function(
                pso_candidate,
                X_data,
                y_data
            )

            current_score = fitness_function(
                population[i],
                X_data,
                y_data
            )

            if pso_score > current_score:
                population[i] = pso_candidate.copy()
                current_score = pso_score

            if current_score > personal_best_score[i]:
                personal_best_score[i] = current_score
                personal_best_position[i] = population[i].copy()

        best_index = np.argmax(personal_best_score)

        if personal_best_score[best_index] > global_best_score:
            global_best_score = personal_best_score[best_index]
            global_best_position = personal_best_position[best_index].copy()

        fitness_history.append(global_best_score)

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best Hybrid CSA+PSO Fitness: {global_best_score:.4f}"
        )

    return global_best_position, global_best_score, fitness_history


selected_mask, best_psa_fitness, psa_history = hybrid_csa_pso_feature_selection(
    X_scaled,
    y
)

selected_feature_indices = np.where(selected_mask == 1)[0]
selected_features = [feature_cols[i] for i in selected_feature_indices]

print("\nSelected Features by Hybrid CSA + PSO:")
print(selected_features)

X_selected = X_scaled[:, selected_feature_indices]


# ============================================================
# Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_selected,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)


# ============================================================
# Hybrid CSA + PSO Optimized Random Forest Model
# ============================================================

psa_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    random_state=42
)

psa_model.fit(X_train, y_train)

y_pred = psa_model.predict(X_test)
psa_accuracy = accuracy_score(y_test, y_pred)

print("\nHybrid CSA + PSO Random Forest Accuracy:", psa_accuracy)


# ============================================================
# H5 Deep Learning Model
# ============================================================

num_classes = len(np.unique(y))

y_train_cat = to_categorical(y_train, num_classes)
y_test_cat = to_categorical(y_test, num_classes)

h5_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

h5_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = h5_model.fit(
    X_train,
    y_train_cat,
    validation_data=(X_test, y_test_cat),
    epochs=50,
    batch_size=8,
    verbose=1
)

h5_loss, h5_accuracy = h5_model.evaluate(X_test, y_test_cat, verbose=0)


# ============================================================
# Result CSV
# ============================================================

result_df = pd.DataFrame([
    {
        "Model": "Hybrid CSA + PSO Random Forest",
        "Accuracy": psa_accuracy,
        "Fitness Score": best_psa_fitness,
        "Selected Features": ", ".join(selected_features)
    },
    {
        "Model": "Hybrid CSA + PSO H5 Neural Network",
        "Accuracy": h5_accuracy,
        "Fitness Score": best_psa_fitness,
        "Selected Features": ", ".join(selected_features)
    }
])

result_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "result.csv")
result_df.to_csv(result_csv_path, index=False)

print("\nResult CSV:")
print(result_df)


# ============================================================
# Prediction CSV
# ============================================================

final_pred_encoded = psa_model.predict(X_selected)

df["predicted_route_category"] = target_encoder.inverse_transform(
    final_pred_encoded
)

prediction_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv")
df.to_csv(prediction_csv_path, index=False)

print("\nPrediction CSV Saved:", prediction_csv_path)


# ============================================================
# Graphs
# ============================================================

accuracy_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")
plt.figure(figsize=(8, 5))
plt.bar(result_df["Model"], result_df["Accuracy"])
plt.title("Hybrid CSA + PSO Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(accuracy_graph_path)
plt.show()

comparison_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")
plt.figure(figsize=(8, 5))
plt.plot(result_df["Model"], result_df["Accuracy"], marker="o")
plt.title("Hybrid CSA + PSO Comparison Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.grid(True)
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(comparison_graph_path)
plt.show()

heatmap_path = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")
corr_df = pd.DataFrame(X_selected, columns=selected_features)
corr_df["target"] = y
corr = corr_df.corr()

plt.figure(figsize=(10, 7))
plt.imshow(corr, aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Hybrid CSA + PSO Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_path)
plt.show()

result_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")
plt.figure(figsize=(8, 5))
plt.barh(result_df["Model"], result_df["Accuracy"])
plt.title("Hybrid CSA + PSO Result Graph")
plt.xlabel("Accuracy")
plt.ylabel("Model")
plt.tight_layout()
plt.savefig(result_graph_path)
plt.show()

prediction_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")
prediction_counts = df["predicted_route_category"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(prediction_counts.index, prediction_counts.values)
plt.title("Hybrid CSA + PSO Prediction Category Distribution")
plt.xlabel("Predicted Route Category")
plt.ylabel("Count")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(prediction_graph_path)
plt.show()

optimization_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "optimization_graph.png")
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(psa_history) + 1), psa_history, marker="o")
plt.title("Hybrid CSA + PSO Optimization Progress Graph")
plt.xlabel("Generation")
plt.ylabel("Best Fitness Score")
plt.grid(True)
plt.tight_layout()
plt.savefig(optimization_graph_path)
plt.show()

training_accuracy_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "training_accuracy_graph.png"
)

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("Hybrid CSA + PSO H5 Training Accuracy Graph")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(training_accuracy_graph_path)
plt.show()

cm = confusion_matrix(y_test, y_pred)

confusion_matrix_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "confusion_matrix_graph.png"
)

plt.figure(figsize=(7, 5))
plt.imshow(cm, interpolation="nearest")
plt.title("Hybrid CSA + PSO Confusion Matrix")
plt.colorbar()

classes = target_encoder.classes_

plt.xticks(np.arange(len(classes)), classes, rotation=45)
plt.yticks(np.arange(len(classes)), classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.savefig(confusion_matrix_graph_path)
plt.show()


# ============================================================
# Save PKL Files
# ============================================================

model_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "model.pkl")
scaler_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "encoders.pkl")
feature_selection_pkl_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "selected_features.pkl"
)

with open(model_pkl_path, "wb") as f:
    pickle.dump(psa_model, f)

with open(scaler_pkl_path, "wb") as f:
    pickle.dump(scaler, f)

with open(encoder_pkl_path, "wb") as f:
    pickle.dump(
        {
            "feature_encoders": label_encoders,
            "target_encoder": target_encoder
        },
        f
    )

with open(feature_selection_pkl_path, "wb") as f:
    pickle.dump(
        {
            "selected_mask": selected_mask,
            "selected_features": selected_features,
            "best_fitness": best_psa_fitness
        },
        f
    )


# ============================================================
# Save H5 Model
# ============================================================

h5_model_path = os.path.join(OUTPUT_DIR, PREFIX + "model.h5")
h5_model.save(h5_model_path)


# ============================================================
# Save YAML Config
# ============================================================

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

config = {
    "project_name": "Agricultural Trade Route Optimization System",
    "author": "Sagnik Patra",
    "algorithm": "Hybrid Clonal Selection Algorithm + Particle Swarm Optimization",
    "prefix": PREFIX,
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "target_column": "trade_route_category",
    "all_features": feature_cols,
    "selected_features": selected_features,
    "psa_parameters": {
        "population_size": 20,
        "generations": 25,
        "clone_rate": 5,
        "mutation_rate": 0.15,
        "replacement_rate": 0.3,
        "w": 0.7,
        "c1": 1.5,
        "c2": 1.5,
        "best_fitness": float(best_psa_fitness)
    },
    "files_generated": {
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "h5_model": h5_model_path,
        "pkl_model": model_pkl_path,
        "pkl_scaler": scaler_pkl_path,
        "pkl_encoders": encoder_pkl_path,
        "pkl_selected_features": feature_selection_pkl_path,
        "accuracy_graph": accuracy_graph_path,
        "comparison_graph": comparison_graph_path,
        "heatmap": heatmap_path,
        "result_graph": result_graph_path,
        "prediction_graph": prediction_graph_path,
        "optimization_graph": optimization_graph_path,
        "training_accuracy_graph": training_accuracy_graph_path,
        "confusion_matrix_graph": confusion_matrix_graph_path
    }
}

with open(yaml_path, "w") as f:
    yaml.dump(config, f, default_flow_style=False)


# ============================================================
# Save JSON Results
# ============================================================

json_path = os.path.join(OUTPUT_DIR, PREFIX + "results.json")

json_results = {
    "project_name": "Agricultural Trade Route Optimization System",
    "author": "Sagnik Patra",
    "algorithm": "Hybrid Clonal Selection Algorithm + Particle Swarm Optimization",
    "prefix": PREFIX,
    "psa_accuracy": float(psa_accuracy),
    "h5_accuracy": float(h5_accuracy),
    "best_psa_fitness": float(best_psa_fitness),
    "selected_features": selected_features,
    "model_results": result_df.to_dict(orient="records"),
    "classification_report": classification_report(
        y_test,
        y_pred,
        target_names=target_encoder.classes_,
        output_dict=True
    ),
    "confusion_matrix": cm.tolist(),
    "total_records": int(len(df)),
    "total_selected_features": int(len(selected_features)),
    "target_classes": list(target_encoder.classes_),
    "saved_files": {
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "h5_model": h5_model_path,
        "pkl_model": model_pkl_path,
        "pkl_scaler": scaler_pkl_path,
        "pkl_encoders": encoder_pkl_path,
        "pkl_selected_features": feature_selection_pkl_path,
        "yaml_config": yaml_path,
        "accuracy_graph": accuracy_graph_path,
        "comparison_graph": comparison_graph_path,
        "heatmap": heatmap_path,
        "result_graph": result_graph_path,
        "prediction_graph": prediction_graph_path,
        "optimization_graph": optimization_graph_path,
        "training_accuracy_graph": training_accuracy_graph_path,
        "confusion_matrix_graph": confusion_matrix_graph_path
    }
}

with open(json_path, "w") as f:
    json.dump(json_results, f, indent=4)


print("\nAll Hybrid CSA + PSO Files Saved Successfully")
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("H5 Model:", h5_model_path)
print("PKL Model:", model_pkl_path)
print("YAML:", yaml_path)
print("JSON:", json_path)
print("PSA Accuracy:", psa_accuracy)
print("H5 Accuracy:", h5_accuracy)
print("Best PSA Fitness:", best_psa_fitness)
```
