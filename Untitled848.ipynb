```python
import os
import json
import yaml
import pickle
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System\megambmarketname_n.xls"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Agricultural Trade Route Optimization System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "woa_"


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_excel(DATASET_PATH)
df.columns = df.columns.str.strip()

required_columns = ["District Name", "Block Name", "Market Name"]

for col in required_columns:
    if col not in df.columns:
        raise ValueError(f"Missing required column: {col}")

df = df.dropna()
df = df.drop_duplicates()

for col in required_columns:
    df[col] = df[col].astype(str).str.strip()


# ============================================================
# Feature Engineering
# ============================================================

df["district_market_count"] = df.groupby("District Name")["Market Name"].transform("count")
df["block_market_count"] = df.groupby("Block Name")["Market Name"].transform("count")

df["market_name_length"] = df["Market Name"].apply(len)
df["block_name_length"] = df["Block Name"].apply(len)
df["district_name_length"] = df["District Name"].apply(len)

df["route_density_score"] = (
    df["district_market_count"] * 0.5 +
    df["block_market_count"] * 0.4 +
    df["market_name_length"] * 0.1
)

df["trade_route_category"] = pd.qcut(
    df["route_density_score"],
    q=3,
    labels=[
        "Low Route Priority",
        "Medium Route Priority",
        "High Route Priority"
    ],
    duplicates="drop"
)


# ============================================================
# Encoding
# ============================================================

label_encoders = {}

for col in required_columns:
    le = LabelEncoder()
    df[col + "_encoded"] = le.fit_transform(df[col])
    label_encoders[col] = le

target_encoder = LabelEncoder()
df["target"] = target_encoder.fit_transform(df["trade_route_category"])


# ============================================================
# Features and Target
# ============================================================

feature_cols = [
    "District Name_encoded",
    "Block Name_encoded",
    "Market Name_encoded",
    "district_market_count",
    "block_market_count",
    "market_name_length",
    "block_name_length",
    "district_name_length",
    "route_density_score"
]

X = df[feature_cols].values
y = df["target"].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# ============================================================
# WOA Feature Selection
# Whale Optimization Algorithm
# ============================================================

def fitness_function(mask, X_data, y_data):
    selected_features = np.where(mask == 1)[0]

    if len(selected_features) == 0:
        return 0

    X_selected = X_data[:, selected_features]

    X_train_fs, X_test_fs, y_train_fs, y_test_fs = train_test_split(
        X_selected,
        y_data,
        test_size=0.2,
        random_state=42,
        stratify=y_data
    )

    model = RandomForestClassifier(
        n_estimators=100,
        max_depth=8,
        random_state=42
    )

    model.fit(X_train_fs, y_train_fs)
    pred = model.predict(X_test_fs)

    accuracy = accuracy_score(y_test_fs, pred)
    feature_penalty = len(selected_features) / X_data.shape[1]

    return accuracy - (0.01 * feature_penalty)


def woa_feature_selection(
    X_data,
    y_data,
    population_size=20,
    generations=25
):
    num_features = X_data.shape[1]

    whales = np.random.rand(population_size, num_features)

    best_position = whales[0].copy()
    best_fitness = -np.inf

    fitness_history = []

    for generation in range(generations):
        a = 2 - generation * (2 / generations)

        for i in range(population_size):
            binary_position = (whales[i] > 0.5).astype(int)

            if np.sum(binary_position) == 0:
                binary_position[np.random.randint(0, num_features)] = 1

            fitness = fitness_function(binary_position, X_data, y_data)

            if fitness > best_fitness:
                best_fitness = fitness
                best_position = whales[i].copy()

        for i in range(population_size):
            r1 = np.random.rand()
            r2 = np.random.rand()

            A = 2 * a * r1 - a
            C = 2 * r2

            p = np.random.rand()

            if p < 0.5:
                if abs(A) < 1:
                    D = np.abs(C * best_position - whales[i])
                    whales[i] = best_position - A * D
                else:
                    random_index = np.random.randint(population_size)
                    random_whale = whales[random_index]

                    D = np.abs(C * random_whale - whales[i])
                    whales[i] = random_whale - A * D
            else:
                b = 1
                l = np.random.uniform(-1, 1)

                D = np.abs(best_position - whales[i])

                whales[i] = (
                    D * np.exp(b * l)
                    * np.cos(2 * np.pi * l)
                    + best_position
                )

        whales = np.clip(whales, 0, 1)

        fitness_history.append(best_fitness)

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best WOA Fitness: {best_fitness:.4f}"
        )

    selected_mask = (best_position > 0.5).astype(int)

    if np.sum(selected_mask) == 0:
        selected_mask[np.random.randint(0, num_features)] = 1

    return selected_mask, best_fitness, fitness_history


selected_mask, best_woa_fitness, woa_history = woa_feature_selection(
    X_scaled,
    y
)

selected_feature_indices = np.where(selected_mask == 1)[0]
selected_features = [feature_cols[i] for i in selected_feature_indices]

print("\nSelected Features by WOA:")
print(selected_features)

X_selected = X_scaled[:, selected_feature_indices]


# ============================================================
# Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_selected,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)


# ============================================================
# WOA Optimized Random Forest Model
# ============================================================

woa_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    random_state=42
)

woa_model.fit(X_train, y_train)

y_pred = woa_model.predict(X_test)
woa_accuracy = accuracy_score(y_test, y_pred)

print("\nWOA Optimized Random Forest Accuracy:", woa_accuracy)


# ============================================================
# H5 Deep Learning Model
# ============================================================

num_classes = len(np.unique(y))

y_train_cat = to_categorical(y_train, num_classes)
y_test_cat = to_categorical(y_test, num_classes)

h5_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

h5_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = h5_model.fit(
    X_train,
    y_train_cat,
    validation_data=(X_test, y_test_cat),
    epochs=50,
    batch_size=8,
    verbose=1
)

h5_loss, h5_accuracy = h5_model.evaluate(X_test, y_test_cat, verbose=0)


# ============================================================
# Result CSV
# ============================================================

result_df = pd.DataFrame([
    {
        "Model": "WOA Optimized Random Forest",
        "Accuracy": woa_accuracy,
        "Fitness Score": best_woa_fitness,
        "Selected Features": ", ".join(selected_features)
    },
    {
        "Model": "WOA Optimized H5 Neural Network",
        "Accuracy": h5_accuracy,
        "Fitness Score": best_woa_fitness,
        "Selected Features": ", ".join(selected_features)
    }
])

result_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "result.csv")
result_df.to_csv(result_csv_path, index=False)

print("\nResult CSV:")
print(result_df)


# ============================================================
# Prediction CSV
# ============================================================

final_pred_encoded = woa_model.predict(X_selected)

df["predicted_route_category"] = target_encoder.inverse_transform(
    final_pred_encoded
)

prediction_csv_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv")
df.to_csv(prediction_csv_path, index=False)

print("\nPrediction CSV Saved:", prediction_csv_path)


# ============================================================
# Graphs
# ============================================================

accuracy_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")
plt.figure(figsize=(8, 5))
plt.bar(result_df["Model"], result_df["Accuracy"])
plt.title("WOA Model Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(accuracy_graph_path)
plt.show()

comparison_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")
plt.figure(figsize=(8, 5))
plt.plot(result_df["Model"], result_df["Accuracy"], marker="o")
plt.title("WOA Model Comparison Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.grid(True)
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(comparison_graph_path)
plt.show()

heatmap_path = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")
corr_df = pd.DataFrame(X_selected, columns=selected_features)
corr_df["target"] = y
corr = corr_df.corr()

plt.figure(figsize=(10, 7))
plt.imshow(corr, aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("WOA Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_path)
plt.show()

result_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")
plt.figure(figsize=(8, 5))
plt.barh(result_df["Model"], result_df["Accuracy"])
plt.title("WOA Result Graph")
plt.xlabel("Accuracy")
plt.ylabel("Model")
plt.tight_layout()
plt.savefig(result_graph_path)
plt.show()

prediction_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")
prediction_counts = df["predicted_route_category"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(prediction_counts.index, prediction_counts.values)
plt.title("WOA Prediction Category Distribution")
plt.xlabel("Predicted Route Category")
plt.ylabel("Count")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(prediction_graph_path)
plt.show()

optimization_graph_path = os.path.join(OUTPUT_DIR, PREFIX + "optimization_graph.png")
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(woa_history) + 1), woa_history, marker="o")
plt.title("WOA Optimization Progress Graph")
plt.xlabel("Generation")
plt.ylabel("Best Fitness Score")
plt.grid(True)
plt.tight_layout()
plt.savefig(optimization_graph_path)
plt.show()

training_accuracy_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "training_accuracy_graph.png"
)

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("WOA H5 Training Accuracy Graph")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(training_accuracy_graph_path)
plt.show()

cm = confusion_matrix(y_test, y_pred)

confusion_matrix_graph_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "confusion_matrix_graph.png"
)

plt.figure(figsize=(7, 5))
plt.imshow(cm, interpolation="nearest")
plt.title("WOA Confusion Matrix")
plt.colorbar()

classes = target_encoder.classes_

plt.xticks(np.arange(len(classes)), classes, rotation=45)
plt.yticks(np.arange(len(classes)), classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.savefig(confusion_matrix_graph_path)
plt.show()


# ============================================================
# Save PKL Files
# ============================================================

model_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "model.pkl")
scaler_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "encoders.pkl")
feature_selection_pkl_path = os.path.join(
    OUTPUT_DIR,
    PREFIX + "selected_features.pkl"
)

with open(model_pkl_path, "wb") as f:
    pickle.dump(woa_model, f)

with open(scaler_pkl_path, "wb") as f:
    pickle.dump(scaler, f)

with open(encoder_pkl_path, "wb") as f:
    pickle.dump(
        {
            "feature_encoders": label_encoders,
            "target_encoder": target_encoder
        },
        f
    )

with open(feature_selection_pkl_path, "wb") as f:
    pickle.dump(
        {
            "selected_mask": selected_mask,
            "selected_features": selected_features,
            "best_fitness": best_woa_fitness
        },
        f
    )


# ============================================================
# Save H5 Model
# ============================================================

h5_model_path = os.path.join(OUTPUT_DIR, PREFIX + "model.h5")
h5_model.save(h5_model_path)


# ============================================================
# Save YAML Config
# ============================================================

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

config = {
    "project_name": "Agricultural Trade Route Optimization System",
    "author": "Sagnik Patra",
    "algorithm": "Whale Optimization Algorithm",
    "prefix": PREFIX,
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "target_column": "trade_route_category",
    "all_features": feature_cols,
    "selected_features": selected_features,
    "woa_parameters": {
        "population_size": 20,
        "generations": 25,
        "best_fitness": float(best_woa_fitness)
    },
    "files_generated": {
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "h5_model": h5_model_path,
        "pkl_model": model_pkl_path,
        "pkl_scaler": scaler_pkl_path,
        "pkl_encoders": encoder_pkl_path,
        "pkl_selected_features": feature_selection_pkl_path,
        "accuracy_graph": accuracy_graph_path,
        "comparison_graph": comparison_graph_path,
        "heatmap": heatmap_path,
        "result_graph": result_graph_path,
        "prediction_graph": prediction_graph_path,
        "optimization_graph": optimization_graph_path,
        "training_accuracy_graph": training_accuracy_graph_path,
        "confusion_matrix_graph": confusion_matrix_graph_path
    }
}

with open(yaml_path, "w") as f:
    yaml.dump(config, f, default_flow_style=False)


# ============================================================
# Save JSON Results
# ============================================================

json_path = os.path.join(OUTPUT_DIR, PREFIX + "results.json")

json_results = {
    "project_name": "Agricultural Trade Route Optimization System",
    "author": "Sagnik Patra",
    "algorithm": "Whale Optimization Algorithm",
    "prefix": PREFIX,
    "woa_accuracy": float(woa_accuracy),
    "h5_accuracy": float(h5_accuracy),
    "best_woa_fitness": float(best_woa_fitness),
    "selected_features": selected_features,
    "model_results": result_df.to_dict(orient="records"),
    "classification_report": classification_report(
        y_test,
        y_pred,
        target_names=target_encoder.classes_,
        output_dict=True
    ),
    "confusion_matrix": cm.tolist(),
    "total_records": int(len(df)),
    "total_selected_features": int(len(selected_features)),
    "target_classes": list(target_encoder.classes_),
    "saved_files": {
        "result_csv": result_csv_path,
        "prediction_csv": prediction_csv_path,
        "h5_model": h5_model_path,
        "pkl_model": model_pkl_path,
        "pkl_scaler": scaler_pkl_path,
        "pkl_encoders": encoder_pkl_path,
        "pkl_selected_features": feature_selection_pkl_path,
        "yaml_config": yaml_path,
        "accuracy_graph": accuracy_graph_path,
        "comparison_graph": comparison_graph_path,
        "heatmap": heatmap_path,
        "result_graph": result_graph_path,
        "prediction_graph": prediction_graph_path,
        "optimization_graph": optimization_graph_path,
        "training_accuracy_graph": training_accuracy_graph_path,
        "confusion_matrix_graph": confusion_matrix_graph_path
    }
}

with open(json_path, "w") as f:
    json.dump(json_results, f, indent=4)


print("\nAll WOA Files Saved Successfully")
print("Result CSV:", result_csv_path)
print("Prediction CSV:", prediction_csv_path)
print("H5 Model:", h5_model_path)
print("PKL Model:", model_pkl_path)
print("YAML:", yaml_path)
print("JSON:", json_path)
print("WOA Accuracy:", woa_accuracy)
print("H5 Accuracy:", h5_accuracy)
print("Best WOA Fitness:", best_woa_fitness)
```
